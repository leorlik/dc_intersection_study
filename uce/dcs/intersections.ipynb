```python
import matplotlib.pyplot as plt
from venny4py.venny4py import venny4py

def read_file_to_set(filepath):
    """Reads lines from a file and returns them as a set."""
    with open(filepath, 'r', encoding='utf-8') as f:
        return {line.strip() for line in f if line.strip()}

# Example usage for four files
set1 = read_file_to_set("all.txt")
set2 = read_file_to_set("pt_1.txt")
set3 = read_file_to_set("pt_2.txt")

data = {
    'Completo': set1,
    'Primeira metade': set2,
    'Segunda metade': set3
}

venny4py(data)
plt.title('Diagrama do Dataset UCE')
plt.show()
```
